```python
#任务4:pdf格式数据的读写查询
import pdfplumber
pdf = pdfplumber.open(R'高考核心词汇1278.pdf')
```

```python
len(pdf.pages)
len([])
```

```python
#初始化列表
table =[]
#从第一页循环到最后一页
for i in range(len(pdf.pages)-1):
    #获取第i页
    page = pdf.pages[i + 1]
    #page.extract_table()作用是提取当前页面page中的表格
    #table.extend的作用是将元素增加到列表table中
    table.extend(page.extract_table())
    
```

```python
pdf_df=pd.DataFrame(table[2:])
```

```python
pdf_df.to_excel('output/pdf表格数据读写练习结果.xlsx',index=False)
```
